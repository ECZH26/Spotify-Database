```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_rows = 50
pd.options.display.max_columns = 999
pd.options.display.float_format = '{:,.4f}'.format
```

```python
spotify = pd.read_csv('Spotify_dataset.csv')
spotify = spotify.drop(columns='Unnamed: 0')
spotify
```

```python
column_name = 'genres'
spotify_onlygenres_cleanup1 = spotify.dropna(subset=['genres'])
spotify_onlygenres = spotify_onlygenres_cleanup1.loc[spotify_onlygenres_cleanup1['genres'] != '[]']
spotify_onlygenres
```

```python
energy_in_decades_all = spotify.groupby('decade').energy.describe()
energy_in_decades_all #orginal 
```

```python
energy_in_decades_nonull = spotify_onlygenres.groupby('decade').energy.describe()
energy_in_decades_nonull #altered
```

```python
nonull_hit = spotify_onlygenres[spotify_onlygenres.target == 1]
nonull_hit # this only contains hit
```

```python
nonull_hitPOP = nonull_hit[nonull_hit.genres.str.contains('pop') & ~nonull_hit.genres.str.contains('rock')] 
# both conditions must be true (&), the ~ negates the second condition
nonull_hitPOP #this only cotains hit song that are in pop genre
```

```python
nonull_hit_decades = nonull_hit.groupby('decade').genres.describe()
nonull_hit_decades
```

```python
nonull_hitPOP_decades = nonull_hitPOP.groupby('decade').genres.describe() # this is just pop
nonull_hitPOP_decades #just deepdiving
```

```python
nonull_hitROCK = nonull_hit[nonull_hit.genres.str.contains('rock') & ~nonull_hit.genres.str.contains('pop')]
# both conditions must be true (&), the ~ negates the second condition
nonull_hitROCK #checking the rock genre
```

```python
nonull_hitROCK_decades = nonull_hitROCK.groupby('decade').genres.describe() # this is just rock
nonull_hitROCK_decades #just deepdiving 
```

```python
nonull_hitDISCO = nonull_hit[nonull_hit.genres.str.contains('disco')]
nonull_hitDISCO_byDecades = nonull_hitDISCO.groupby('decade').target.count()
nonull_hitDISCO_byDecades
```

```python
hits_by_decade = nonull_hit.groupby('decade').target.count() # total hit songs with no nulls
hits_by_decade
```

```python
nonull_hitPOP_byDECADE = nonull_hitPOP_decades['count'] #total of hit pop song produce no null
nonull_hitPOP_byDECADE
```

```python
hit_pop_songs_percentage = nonull_hitPOP_byDECADE/hits_by_decade*100 # percentage of hit pop song produced
hit_pop_songs_percentage
```

```python
nonull_hitROCK_byDECADE = nonull_hitROCK_decades['count'] #total of hit rock song produce no null
nonull_hitROCK_byDECADE
```

```python
hit_rock_songs_percentage = nonull_hitROCK_byDECADE/hits_by_decade*100 # percentage of rock song produced
hit_rock_songs_percentage
```

```python
x = ['1960s', '1970s', '1980s', '1990s', '2000s', '2010s']
fig, ax = plt.subplots()
ax.plot(x, nonull_hitPOP_byDECADE, label='Pop Count per Decade')
ax.plot(x, nonull_hitROCK_byDECADE, label='Rock Count per Decade')
ax.plot(x, hits_by_decade, label='Total Hit Count per Decade')
ax.set_xlabel('Decades')
ax.set_ylabel('Song Count')
ax.set_title('Hits per Decade')
ax.legend()
plt.show()
```

```python
x = ['1960s', '1970s', '1980s', '1990s', '2000s', '2010s']
# calculate total non-null hits for each decade by adding up the pop and rock counts
total_nonull_hits = [sum(x) for x in zip(nonull_hitPOP_byDECADE, nonull_hitROCK_byDECADE)]
# create a figure and axis object
fig, ax = plt.subplots()
# all graphs will be line graphs
# plot the pop count per decade
ax.plot(x, nonull_hitPOP_byDECADE, label='Pop Count per Decade')
# plot the rock count per decade
ax.plot(x, nonull_hitROCK_byDECADE, label='Rock Count per Decade')
# plot the total hit count per decade
ax.plot(x, hits_by_decade, label='Total Hit Count per Decade')
# plot the combined rock and pop count per decade
ax.plot(x, total_nonull_hits, label='Rock and Pop Songs Combined Count per Decade')
# set the axis labels
ax.set_xlabel('Decades', fontsize=12)
ax.set_ylabel('Song Count', fontsize=12)
# set the title of the plot
ax.set_title('Hits per Decade', fontsize=14)
# add a legend to the plot with smaller font size than normal and display the plot
ax.legend(fontsize=7)
plt.show()
```

```python
# define the x-axis values as a list of strings representing decades
x = ['1960s', '1970s', '1980s', '1990s', '2000s', '2010s']
# create a figure and axis object using subplots function
fig, ax = plt.subplots()
# define the width of the bars in the histogram
bar_width = 0.4
# create a list of bar positions using a for loop
bar_positions = []
for i in range(len(x)):
    bar_positions.append(i)
# create a histogram bar for the 'Rock' songs using the bar function, and shift each bar to the left
ax.bar([i - bar_width/2 for i in bar_positions], nonull_hitROCK_byDECADE, width=bar_width, label='Rock')
# create a histogram bar for the 'Pop' songs using the bar function, and shift each bar to the right
ax.bar([i + bar_width/2 for i in bar_positions], nonull_hitPOP_byDECADE, width=bar_width, label='Pop')
# set the axis labels and title
ax.set_xlabel('Decades')
ax.set_ylabel('Song Count')
ax.set_title('Rock and Pop Songs Non-Null Hit Count per Decade')
# set the x-ticks to be the bar positions, and label each tick with the corresponding decade
ax.set_xticks(bar_positions)
ax.set_xticklabels(x)
# add a legend to the histogram and display it
ax.legend()
plt.show()
```

```python
x = ['1960s', '1970s', '1980s', '1990s', '2000s', '2010s']

fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))

colors = ['#1f77b4', '#d62728']  # blue for pop songs, red for rock songs

for i, decade in enumerate(x):

    pop_size = nonull_hitPOP_byDECADE[i]
    rock_size = nonull_hitROCK_byDECADE[i]

    row = i // 3
    col = i % 3
    axs[row, col].pie([pop_size, rock_size], labels=['Hit Pop Songs', 'Hit Rock Songs'], colors=colors, autopct='%1.2f%%')
    axs[row, col].set_title(decade)
fig.suptitle('Percentage of Hit Pop Songs vs Hit Rock Songs per Decade', fontsize=14)
plt.show()
```

```python
flops_and_hits = spotify.groupby(['decade', 'target'])['target'].count().unstack()
flops_and_hits.loc['Total Count'] = flops_and_hits.sum()
fig, ax = plt.subplots()
ax.axis('off')
ax.axis('tight')
ax.table(cellText=flops_and_hits.values,
         colLabels=['Flops', 'Hits'],
         rowLabels=flops_and_hits.index,
         loc='center')
ax.set_title('Flops and Hits Counts by Decade')
fig.tight_layout()
plt.show()
```

```python
flops_and_hits2 = spotify_onlygenres.groupby(['decade', 'target'])['target'].count().unstack()
flops_and_hits2.loc['Total Count'] = flops_and_hits2.sum()
fig, ax = plt.subplots()
ax.axis('off')
ax.axis('tight')
ax.table(cellText=flops_and_hits2.values,
         colLabels=['Flops', 'Hits'],
         rowLabels=flops_and_hits2.index,
         loc='center')
ax.set_title('Flops and Hits Counts by Decade w/ no Nulls')
fig.tight_layout()
plt.show()
```

```python

total_counts = [flops_and_hits.loc['Total Count'], flops_and_hits2.loc['Total Count']]

diff = flops_and_hits2.sub(flops_and_hits).add_prefix('difference ')

combined_table = pd.concat([flops_and_hits, flops_and_hits2, diff], axis=1, sort=False)
combined_table.columns = ['Non-Modified Flops', 'Non-Modified Hits', 'Modified Flops', 'Modified Hits', 'Difference Between Flops', 'Difference Between Hits']

fig, ax = plt.subplots(figsize=(10, 10))
ax.axis('off')
ax.axis('tight')
ax.table(cellText=combined_table.values,
         colLabels=combined_table.columns,
         rowLabels=combined_table.index,
         loc='center')
ax.set_title('Flops and Hits Counts by Decade', fontsize=10)
fig.tight_layout()
plt.show()

# FIX SCALING TOMORROW
```
